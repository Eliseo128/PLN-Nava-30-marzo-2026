```python
import time

def mostrar_bienvenida():
    print("------------------------------------------------")
    print("🤖 ¡Bienvenido al Asistente Virtual del Colegio!")
    print("------------------------------------------------")

def mostrar_horario():
    print("\n📅 Aquí tienes tu horario de esta semana:")
    print("Lunes: Matemáticas, Historia, Educación Física")
    print("Martes: Ciencias, Lengua, Arte")
    print("Miércoles: Matemáticas, Inglés, Tecnología")
    print("Jueves: Historia, Ciencias, Música")
    print("Viernes: Lengua, Tutoría, Deportes")

def buscar_aula():
    numero_aula = input("🏫 ¿Qué número de aula buscas? (ej. 101, 205): ")

    # Lógica de decisión simple basada en el número
    if numero_aula.startswith("1"):
        print(f"✅ El aula {numero_aula} está en la PLANTA BAJA (Pasillo A).")
    elif numero_aula.startswith("2"):
        print(f"✅ El aula {numero_aula} está en la PRIMERA PLANTA (Pasillo B).")
    elif numero_aula.startswith("3"):
        print(f"✅ El aula {numero_aula} está en la SEGUNDA PLANTA (Laboratorios).")
    else:
        print("❌ No reconozco ese número de aula. Consulta en secretaría.")

def main():
    mostrar_bienvenida()

    # Bucle infinito para mantener el asistente activo
    while True:
        print("\n¿En qué puedo ayudarte hoy?")
        print("1. Ver Horario")
        print("2. Buscar Aula")
        print("3. Salir")

        opcion = input("Escribe el número de la opción (1, 2 o 3): ")

        print("\n⏳ Procesando solicitud...")
        time.sleep(1) # Simula que el robot está "pensando"

        if opcion == "1":
            mostrar_horario()
        elif opcion == "2":
            buscar_aula()
        elif opcion == "3":
            print("\n👋 ¡Hasta luego! Que tengas un buen día de estudio.")
            break # Rompe el bucle y termina el programa
        else:
            print("⚠️ Opción no válida. Por favor, elige 1, 2 o 3.")

# Ejecutar el programa
if __name__ == "__main__":
    main()
```

```python
import time

def verificar_pin():
    """Simula la seguridad del cajero pidiendo un PIN."""
    pin_correcto = "1234"
    intentos = 3

    while intentos > 0:
        pin_usuario = input("🔒 Ingresa tu PIN (4 dígitos): ")
        if pin_usuario == pin_correcto:
            print("✅ Acceso concedido.\n")
            return True
        else:
            intentos -= 1
            print(f"❌ PIN incorrecto. Te quedan {intentos} intentos.")

    print("🚫 Tarjeta bloqueada por seguridad.")
    return False

def mostrar_menu():
    print("----------- MENÚ ATM -----------")
    print("1. Consultar Saldo")
    print("2. Depositar Dinero")
    print("3. Retirar Dinero")
    print("4. Salir")
    print("--------------------------------")

def main():
    # Variable de estado: El saldo inicial del cliente
    saldo_actual = 1000.00

    print("🏧 Bienvenido al Cajero Automático PythonBank")

    # Paso 1: Seguridad
    if not verificar_pin():
        return # Termina el programa si el PIN falla

    # Paso 2: Bucle principal de operaciones
    while True:
        mostrar_menu()
        opcion = input("Selecciona una opción (1-4): ")

        print("\n⏳ Conectando con el banco...")
        time.sleep(1)

        if opcion == "1":
            # Formateo de moneda para que se vea profesional (2 decimales)
            print(f"💰 Tu saldo actual es: ${saldo_actual:.2f}")

        elif opcion == "2":
            try:
                cantidad = float(input("📥 ¿Cuánto dinero deseas depositar?: $"))
                if cantidad > 0:
                    saldo_actual += cantidad
                    print(f"✅ Depósito exitoso. Nuevo saldo: ${saldo_actual:.2f}")
                else:
                    print("❌ No puedes depositar una cantidad negativa o cero.")
            except ValueError:
                print("❌ Error: Por favor ingresa un número válido.")

        elif opcion == "3":
            try:
                cantidad = float(input("📤 ¿Cuánto dinero deseas retirar?: $"))
                if cantidad > 0:
                    if cantidad <= saldo_actual:
                        saldo_actual -= cantidad
                        print(f"✅ Retiro exitoso. Nuevo saldo: ${saldo_actual:.2f}")
                    else:
                        print("❌ Fondos insuficientes para realizar esta operación.")
                else:
                    print("❌ No puedes retirar una cantidad negativa o cero.")
            except ValueError:
                print("❌ Error: Por favor ingresa un número válido.")

        elif opcion == "4":
            print("\n👋 Gracias por usar PythonBank. ¡Hasta pronto!")
            break # Sale del bucle while

        else:
            print("⚠️ Opción no válida. Intente de nuevo.")

        # Pequeña pausa para limpiar la pantalla visualmente (opcional)
        print("\n")

# Ejecutar el programa
if __name__ == "__main__":
    main()
```
